### Libraries

```python
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
```

### Visualizations

```python
def Lineplot(df, x, y, hue=None):
    plt.figure(figsize=(12, 6))
    sns.lineplot(x=x, y=y, hue=hue, data=df, marker='o', markersize=10)
    plt.title(f'{y} by {x}', fontweight="bold", fontfamily="Arial")
    sns.despine(top=True, right=True, left=True, bottom=True)
    plt.grid(True)
    plt.show()
```

### Data

```python
training_data = pd.DataFrame({ 
    'iterations': [20, 400, 2000], 
    'training time in seconds': [63, 1236, 6337] 
    })


training_data["seconds per iteration"] = training_data["training time in seconds"] / training_data["iterations"]
```

```python
Lineplot(training_data, 'iterations', 'training time in seconds')
```
